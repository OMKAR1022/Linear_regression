```python
import pandas as pd
import math
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
df = pd.read_csv('c_housing.csv')
```

```python
df
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
bed_median = math.floor(df['total_bedrooms'].mean())
bed_median
```

```python
df['total_bedrooms'] = df['total_bedrooms'].fillna(bed_median)
df
```

```python
df.isnull().sum()
```

```python
label_encoder = LabelEncoder()
```

```python
df_label_encoder = df
```

```python
df_label_encoder['ocean_proximity'] = label_encoder.fit_transform(df_label_encoder['ocean_proximity'])
```

```python
df_label_encoder
```

```python
X = df_label_encoder.drop(['median_house_value'],axis=1).values
```

```python
X
```

```python
y = df_label_encoder.median_house_value.values
```

```python
y
```

```python
Column_Transforme = ColumnTransformer([('ocean_proximity',OneHotEncoder(),[8])],remainder="passthrough")
```

```python
Column_Transforme
```

```python
X = Column_Transforme.fit_transform(X)
X
```

```python
X= np.array(X)
X
```

```python
X = np.delete(X,7,axis=1)
```

```python
X
```

```python
model = LinearRegression()
```

```python
X_train ,X_test , y_train , y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
