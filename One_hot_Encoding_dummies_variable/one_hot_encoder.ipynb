```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('carprices.csv')
```

```python
df.head(12)
```

```python
label_encoder = LabelEncoder()
```

```python
df_label_encoder = df
```

```python
df_label_encoder['Car Model'] = label_encoder.fit_transform(df_label_encoder['Car Model'])
```

```python
df_label_encoder
```

```python
X = df_label_encoder[['Car Model','Mileage','Age(yrs)']].values
```

```python
X
```

```python
y = df_label_encoder['Sell Price($)'].values
```

```python
y
```

```python
column_transformer = ColumnTransformer([('Car Model', OneHotEncoder() ,[0])], remainder='passthrough')
```

```python
column_transformer
```

```python
X = column_transformer.fit_transform(X)
```

```python
X
```

# Remove one dummies column

```python
X = X[:,1:]
```

```python
X
```

```python
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[0,1,45000,4]])
```
